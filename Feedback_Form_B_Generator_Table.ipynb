```python
!pip install python-docx

import requests
from docx import Document
import re

# Download the template from GitHub
url = 'https://raw.githubusercontent.com/HistoriansGulyas/CWS/main/Item%20B%20Feedback%20Form-Table.docx'
template_path = '/content/Item B Feedback Form-Table.docx'

try:
    print(f"Downloading template from {url}...")
    response = requests.get(url)
    response.raise_for_status() # Check for HTTP errors
    with open(template_path, 'wb') as f:
        f.write(response.content)
    print("Template downloaded successfully.")

    # Load the document to inspect placeholders
    doc = Document(template_path)

    # Initialize a set to store unique placeholders
    placeholders = set()

    # Regex pattern to find text inside square brackets
    pattern = r'\[.*?\]'

    # Iterate through paragraphs in the document body
    for paragraph in doc.paragraphs:
        matches = re.findall(pattern, paragraph.text)
        placeholders.update(matches)

    # Iterate through tables in the document
    for table in doc.tables:
        for row in table.rows:
            for cell in row.cells:
                for paragraph in cell.paragraphs:
                    matches = re.findall(pattern, paragraph.text)
                    placeholders.update(matches)

    # Sort and print the unique placeholders found
    sorted_placeholders = sorted(list(placeholders))
    print("Unique placeholders found in the template:")
    for p in sorted_placeholders:
        print(p)

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
satisfaction_levels = ['very satisfied', 'satisfied', 'neutral', 'unsatisfied', 'very unsatisfied']

def get_counts(df, col_name):
    """Counts occurrences of each rating category."""
    if col_name not in df.columns:
        return {level: 0 for level in satisfaction_levels}
    counts = df[col_name].value_counts()
    return {level: counts.get(level, 0) for level in satisfaction_levels}

# Dictionary to store all processed data
processed_data = {
    'LO_COUNTS': [],
    'TRAINER_COUNTS': [],
    'CONSULTANT_COUNTS': [],
    'OTHER_COUNTS': {},
    'COMMENTS': {}
}

# 1. Learning Objectives
lo_questions = {
    1: "How well do you feel you can summarize the EMDR approach to psychotherapy and integrate the EMDR methodology?",
    2: "How well do you feel you can summarize an overview of traumatology and neurobiology of trauma?",
    3: "How well do you feel you can integrate and apply standard EMDR Therapy to psychotherapy?",
    4: "How well do you feel you practiced the basic standard EMDR protocol to feel competent?",
    5: "How well do you feel you can apply EMDR Therapy with appropriate existing and new clients?",
    6: "LO 6",
    7: "LO 7"
}

for i in range(1, 8):
    col = f'Learning Objectives {i}'
    counts = get_counts(df, col)
    # Use the specific question text, defaulting to 'LO i' if not found
    question_text = lo_questions.get(i, f'LO {i}')
    processed_data['LO_COUNTS'].append({'Question': question_text, **counts})

# 2. Trainers
for name in trainers:
    # Find rating column (fuzzy match or exact construction)
    rating_col = next((c for c in df.columns if name in c and 'Rating' in c), None)
    if rating_col:
        counts = get_counts(df, rating_col)
        processed_data['TRAINER_COUNTS'].append({'Name': name, **counts})

    # Aggregate Comments
    comment_col = next((c for c in df.columns if name in c and 'Comments' in c), None)
    if comment_col:
        comments = df[comment_col].dropna().astype(str).tolist()
        processed_data['COMMENTS'][f'{name}_COMMENTS'] = "\n\n---\n\n".join(comments)

# 3. Consultants
for name in consultants:
    if not name: continue
    rating_col = next((c for c in df.columns if name in c and 'Rating' in c), None)
    if rating_col:
        counts = get_counts(df, rating_col)
        processed_data['CONSULTANT_COUNTS'].append({'Name': name, **counts})

    comment_col = next((c for c in df.columns if name in c and 'Comments' in c), None)
    if comment_col:
        comments = df[comment_col].dropna().astype(str).tolist()
        processed_data['COMMENTS'][f'{name}_COMMENTS'] = "\n\n---\n\n".join(comments)

# 4. Other Tables (OT, CWS, HT)
other_mappings = [
    ('Overall Training Rating:', 'OT'),
    ('CompassionWorks.com', 'CWS'),
    ('Hybrid Training Format:', 'HT')
]

for col, key in other_mappings:
    processed_data['OTHER_COUNTS'][key] = get_counts(df, col)

# 5. Aggregate General Comments
general_comment_cols = {
    'Overall Training Comments': 'OT_COMMENTS',
    'Hybrid Training Comments:': 'HT_COMMENTS',
    'CompassionWorks.com Comments:': 'CWS_COMMENTS',
    'Learning Objectives Comments:': 'LO_COMMENTS'
}

for col, key in general_comment_cols.items():
    if col in df.columns:
        comments = df[col].dropna().astype(str).tolist()
        processed_data['COMMENTS'][key] = "\n\n---\n\n".join(comments)
    else:
        processed_data['COMMENTS'][key] = ""

# Display summary for verification
print("Processing Complete. Summary of counts:")
print(f"Learning Objectives processed: {len(processed_data['LO_COUNTS'])}")
print(f"Trainers processed: {len(processed_data['TRAINER_COUNTS'])}")
print(f"Consultants processed: {len(processed_data['CONSULTANT_COUNTS'])}")
print("\nExample LO Counts (LO 1):", processed_data['LO_COUNTS'][0])
if processed_data['TRAINER_COUNTS']:
    print("Example Trainer Counts:", processed_data['TRAINER_COUNTS'][0])
```

```python
from docx import Document
from docx.shared import Pt, Inches
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.enum.table import WD_TABLE_ALIGNMENT, WD_ROW_HEIGHT_RULE
from docx.oxml import OxmlElement
from docx.oxml.ns import qn

# Helper function to enforce table borders
def set_table_borders(table):
    """
    Explicitly sets table borders to single lines to ensure gridlines are visible.
    """
    tbl = table._tbl
    tblPr = tbl.tblPr

    # Create a new tblBorders element
    tblBorders = OxmlElement('w:tblBorders')

    # Define specific borders
    for border_name in ['top', 'left', 'bottom', 'right', 'insideH', 'insideV']:
        border = OxmlElement(f'w:{border_name}')
        border.set(qn('w:val'), 'single')
        border.set(qn('w:sz'), '4') # 4 eighths of a point = 1/2 pt
        border.set(qn('w:space'), '0')
        border.set(qn('w:color'), 'auto')
        tblBorders.append(border)

    # Remove existing tblBorders if present to avoid duplicates/conflicts
    current_borders = tblPr.find(qn('w:tblBorders'))
    if current_borders is not None:
        tblPr.remove(current_borders)

    # Append the new borders
    tblPr.append(tblBorders)

# 1. Define Text Replacement Function
def replace_text_in_doc(doc, replacements):
    """Replaces keys in replacements dict with values in all paragraphs and tables."""
    for p in doc.paragraphs:
        for key, val in replacements.items():
            if key in p.text:
                p.text = p.text.replace(key, str(val))

    for table in doc.tables:
        for row in table.rows:
            for cell in row.cells:
                for p in cell.paragraphs:
                    for key, val in replacements.items():
                        if key in p.text:
                            p.text = p.text.replace(key, str(val))

# 2. Define Generic Table Replacement Function
def replace_placeholder_with_table(doc, placeholder, data_list, headers, keys, skip_initial_header=False, autofit_table=False, first_col_width=None, max_row_height=None):
    """
    Finds the paragraph with the placeholder, inserts a table with data_list after it,
    and removes the placeholder paragraph.

    :param doc: The docx Document object
    :param placeholder: The string to find (e.g., '[TABLE]')
    :param data_list: List of dictionaries containing the data
    :param headers: List of strings for the table header row
    :param keys: List of keys to extract data from dicts for each column, corresponding to headers
    :param skip_initial_header: If True, does not create the header row at the top of the table.
    :param autofit_table: If True, allows table to size to content (does not force full width).
    :param first_col_width: If provided (e.g. Inches(1.75)), sets the width of the first column.
    :param max_row_height: If provided (e.g. Inches(1)), sets the row height to this value using EXACTLY rule.
    """
    target_p = None
    # Find the paragraph containing the placeholder
    for p in doc.paragraphs:
        if placeholder in p.text:
            target_p = p
            break

    if target_p:
        # Determine available width from section properties (assuming first section)
        section = doc.sections[0]
        available_width = section.page_width - section.left_margin - section.right_margin

        # Create a new table
        if not data_list:
            # Handle empty data gracefully
            table = doc.add_table(rows=1, cols=len(headers))
            table.style = 'Table Grid'
            table.alignment = WD_TABLE_ALIGNMENT.CENTER
            table.allow_autofit = True # Enable Autofit

            # Only set width if NOT autofit_table
            if not autofit_table:
                table.width = available_width

            set_table_borders(table) # Force borders

            # Apply first column width if specified
            if first_col_width:
                try:
                    table.columns[0].width = first_col_width
                    for cell in table.columns[0].cells:
                        cell.width = first_col_width
                except Exception:
                    pass # Handle potential issues with merged cells gracefully

            hdr_cells = table.rows[0].cells
            for i, h in enumerate(headers):
                hdr_cells[i].text = h
            table.add_row().cells[0].text = "No data available"
        else:
            # Initialize table when data exists
            if skip_initial_header:
                # Start with 0 rows, headers will be added via data_list logic
                table = doc.add_table(rows=0, cols=len(headers))
            else:
                # Standard table with initial header row
                table = doc.add_table(rows=1, cols=len(headers))
                hdr_cells = table.rows[0].cells
                for i, h in enumerate(headers):
                    hdr_cells[i].text = h
                    if hdr_cells[i].paragraphs:
                        hdr_cells[i].paragraphs[0].runs[0].bold = True

            table.style = 'Table Grid'

            # Formatting: Center
            table.alignment = WD_TABLE_ALIGNMENT.CENTER
            table.allow_autofit = True # Enable Autofit

            # Only set width if NOT autofit_table
            if not autofit_table:
                table.allow_autofit = False
                table.width = available_width

            set_table_borders(table) # Force borders

            # Apply first column width if specified
            if first_col_width:
                try:
                    table.columns[0].width = first_col_width
                    for cell in table.columns[0].cells:
                        cell.width = first_col_width

                    # If autofit is False and we have exactly 2 columns, set the 2nd column to the remaining width
                    if not autofit_table and len(headers) == 2:
                        remaining_width = available_width - first_col_width
                        table.columns[1].width = remaining_width
                        for cell in table.columns[1].cells:
                            cell.width = remaining_width

                except Exception:
                    pass # Handle potential issues with merged cells gracefully

            # Populate Data Rows
            for item in data_list:
                row = table.add_row()
                row_cells = row.cells

                # Identify special rows
                is_subheader = 'subheader' in item
                is_header = 'is_header' in item

                # Apply row height if specified AND NOT a special row (header/subheader)
                if max_row_height and not is_subheader and not is_header:
                    row.height_rule = WD_ROW_HEIGHT_RULE.EXACTLY
                    row.height = max_row_height

                # Check for Subheader
                if is_subheader:
                    # Merge all cells in this row
                    merged_cell = row_cells[0].merge(row_cells[-1])
                    merged_cell.text = str(item['subheader'])
                    # Make bold
                    if merged_cell.paragraphs:
                        merged_cell.paragraphs[0].runs[0].bold = True

                # Check for Repeater Header
                elif is_header:
                    for i, h in enumerate(headers):
                        row_cells[i].text = h
                        if row_cells[i].paragraphs:
                            row_cells[i].paragraphs[0].runs[0].bold = True

                else:
                    # Normal Row
                    for i, key in enumerate(keys):
                        cell = row_cells[i]
                        val_text = str(item.get(key, ''))

                        # Logic to bold Name in Question column
                        if key == 'Question' and 'Name' in item and item['Name'] in val_text:
                            cell.text = "" # Clear default
                            p = cell.paragraphs[0]
                            name = item['Name']
                            parts = val_text.split(name)
                            for idx, part in enumerate(parts):
                                p.add_run(part)
                                if idx < len(parts) - 1:
                                    p.add_run(name).bold = True
                        else:
                            cell.text = val_text

                        if cell.paragraphs:
                            # Center align number columns in AGG_TABLE (columns > 0)
                            if placeholder == '[AGG_TABLE]' and i > 0:
                                cell.paragraphs[0].alignment = WD_ALIGN_PARAGRAPH.CENTER
                            else:
                                cell.paragraphs[0].alignment = WD_ALIGN_PARAGRAPH.LEFT

        # Move the table to the position of the target paragraph
        parent = target_p._element.getparent()
        parent.insert(parent.index(target_p._element) + 1, table._element)

        # Remove the original paragraph
        parent.remove(target_p._element)
    else:
        print(f"Placeholder '{placeholder}' not found in document body.")

# Helper to split joined comments back into list
def extract_comments(key):
    raw = processed_data['COMMENTS'].get(key, "")
    if not raw:
        return []
    # Split by the delimiter we used earlier
    return [{'Comment': c.strip()} for c in raw.split('\n\n---\n\n') if c.strip()]

# 3. Load Template
try:
    doc = Document(template_path)
    print(f"Loaded template from {template_path}")

    # 4. Prepare Text Replacements (Metadata only)

    # Join non-empty trainer/consultant names with commas
    trainers_str = ", ".join([t for t in trainers if t])
    consultants_str = ", ".join([c for c in consultants if c])

    text_replacements = {
        '[LOCATION]': location,
        '[NUM_RESP]': str(num_resp),
        '[RESP_RATE]': f"{resp_rate:.2f}%",
        '[TRAINERS]': trainers_str,
        '[CONSULTANTS]': consultants_str
    }

    # Add Trainers (handle up to 3) for individual placeholders if still needed
    for i in range(3):
        key = f'[TRAINER_{i+1}]'
        val = trainers[i] if i < len(trainers) else ""
        text_replacements[key] = val

    # Add Consultants (handle up to 2) for individual placeholders if still needed
    for i in range(2):
        key = f'[CONSULTANT_{i+1}]'
        val = consultants[i] if i < len(consultants) else ""
        text_replacements[key] = val

    # Execute Text Replacements
    replace_text_in_doc(doc, text_replacements)
    print("Text replacements completed.")

    # 5. Prepare Aggregate Count Data ([AGG_TABLE])
    agg_data = []

    # Add Trainers (Map 'Name' to specific Question)
    if processed_data['TRAINER_COUNTS']:
        agg_data.append({'subheader': 'TRAINERS'})
        agg_data.append({'is_header': True})
        for t in processed_data['TRAINER_COUNTS']:
            row = t.copy()
            row['Question'] = f"How would you rate {row['Name']} on Preparedness, Responsiveness to Questions, Concepts Clearly Explained, Professionalism?"
            agg_data.append(row)

    # Add Consultants
    if processed_data['CONSULTANT_COUNTS']:
        agg_data.append({'subheader': 'CONSULTANTS'})
        agg_data.append({'is_header': True})
        for c in processed_data['CONSULTANT_COUNTS']:
            row = c.copy()
            row['Question'] = f"How would you rate {row['Name']} on Preparedness, Responsiveness to Questions, Concepts Clearly Explained, Professionalism?"
            agg_data.append(row)

    # Add Learning Objectives
    if processed_data['LO_COUNTS']:
        agg_data.append({'subheader': 'LEARNING OBJECTIVES'})
        agg_data.append({'is_header': True})
        agg_data.extend(processed_data['LO_COUNTS'])

    # Add Other Metrics
    agg_data.append({'subheader': 'OTHER METRICS'})
    agg_data.append({'is_header': True})
    agg_data.append({'Question': 'Overall Training', **processed_data['OTHER_COUNTS']['OT']})
    agg_data.append({'Question': 'CompassionWorks', **processed_data['OTHER_COUNTS']['CWS']})
    agg_data.append({'Question': 'Hybrid Format', **processed_data['OTHER_COUNTS']['HT']})

    # Headers and Keys for Agg Table
    agg_headers = ['Item/Question', 'Very Satisfied', 'Satisfied', 'Neutral', 'Unsatisfied', 'Very Unsatisfied']
    agg_keys = ['Question', 'very satisfied', 'satisfied', 'neutral', 'unsatisfied', 'very unsatisfied']

    # Replace [AGG_TABLE] - Skip initial header because we add them dynamically
    replace_placeholder_with_table(doc, '[AGG_TABLE]', agg_data, agg_headers, agg_keys, skip_initial_header=True, first_col_width=Inches(1.75))

    # 6. Prepare and Replace Comment Tables

    # Overall Training Comments
    ot_comments = extract_comments('OT_COMMENTS')
    replace_placeholder_with_table(doc, '[OT_TABLE]', ot_comments, ['Overall Training Comments'], ['Comment'], autofit_table=True)

    # CompassionWorks Comments
    cws_comments = extract_comments('CWS_COMMENTS')
    replace_placeholder_with_table(doc, '[CWS_TABLE]', cws_comments, ['CompassionWorks Comments'], ['Comment'], autofit_table=True)

    # Hybrid Format Comments
    ht_comments = extract_comments('HT_COMMENTS')
    replace_placeholder_with_table(doc, '[HT_TABLE]', ht_comments, ['Hybrid Format Comments'], ['Comment'], autofit_table=True)

    # Trainer Comments (Consolidated Table: Name | Comment)
    trainer_comments = []
    for name in trainers:
        if name:
            raw = processed_data['COMMENTS'].get(f'{name}_COMMENTS', "")
            if raw:
                for c in raw.split('\n\n---\n\n'):
                    if c.strip():
                        trainer_comments.append({'Name': name, 'Comment': c.strip()})

    replace_placeholder_with_table(doc, '[TRAINER_TABLE]', trainer_comments, ['Trainer', 'Comments'], ['Name', 'Comment'], autofit_table=False, first_col_width=Inches(1.75))

    # Consultant Comments (Consolidated Table: Name | Comment)
    consultant_comments = []
    for name in consultants:
        if name:
            raw = processed_data['COMMENTS'].get(f'{name}_COMMENTS', "")
            if raw:
                for c in raw.split('\n\n---\n\n'):
                    if c.strip():
                        consultant_comments.append({'Name': name, 'Comment': c.strip()})

    replace_placeholder_with_table(doc, '[CONSULTANT_TABLE]', consultant_comments, ['Consultant', 'Comments'], ['Name', 'Comment'], autofit_table=False, first_col_width=Inches(1.75))

    print("Table replacements completed.")

    # 7. Save Document
    safe_location = "".join([c for c in location if c.isalnum() or c in (' ', '_', '-')]).strip()
    output_filename = f'{safe_location}_Feedback_Report.docx'
    doc.save(output_filename)
    print(f"\nSuccess! Report saved as: {output_filename}")

except Exception as e:
    print(f"An error occurred during report generation: {e}")
```
